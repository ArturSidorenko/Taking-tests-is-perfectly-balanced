```python
import pyautogui
import pandas as pd
import time
from bs4 import BeautifulSoup
import cv2
import pyperclip
import re
```

```python
def read_file(filename):
    with open(filename, encoding='utf-8') as input_file:
        text = input_file.read()
    return text

def parse(t):
    t = read_file('tests.html')
    u = BeautifulSoup(t)
    lig = u.find_all(['h3','strong'])
    lique = u.find_all(['h3'])
    li2 = []
    lique2=[]
    for item in lig:
        s = item.text[3:]
        li2.append(clean_str(s))
    for item in lique:
        s = item.text[3:]
        lique2.append(clean_str(s))
    dic={}
    li3=[]
    q=""
    for item in li2:
        if item in lique2:
            dic[q]=li3
            q = item
            if item in dic.keys():
                li3 = dic[item]
            else:
                li3 = []
        else:
            li3.append(item)
    prev = []
    if q in dic.keys():
        prev = dic[q]
    dic[q] = prev+li3
    return dic

def find_ans(filename):
    t = read_file(filename)
    return parse(t)

def clean_str(data):
    data = data.lower()
    bad_characters = [';', ':', ' ', '+', '\r', '\n', '-', 
                     '.', '?', ',', '\t', '_', ]
    for k in bad_characters:
        data = data.replace(k, '')
    
    #pattern = re.compile('^[a-z, а-я, 0-9]*')
    #result = pattern.findall(data)

    return data
```

```python
def get_coord(boxy):
    return (boxy[0] + 30, boxy[1]+5) 

def get_buttoms():
    boxes = list(pyautogui.locateAllOnScreen(
        'box.png',confidence=0.8, grayscale=True))
    radios = list(pyautogui.locateAllOnScreen(
        'radio.png',confidence=0.8,grayscale=True))
    radios_ticked = list(pyautogui.locateAllOnScreen(
        'radio_ticked.png',confidence=0.8,grayscale=True))
    return boxes+radios+radios_ticked

def take_clip(i):
    pyautogui.moveTo(get_coord(i))
    pyautogui.click(clicks=3, interval=0.3)
    pyautogui.hotkey('ctrl', 'c')
    data = pyperclip.paste()
    
    pyautogui.moveTo(get_coord(i))
    pyautogui.click(clicks=1, interval=0.3)
    
    data = clean_str(data)
    
    return data

def get_question():
    p=list(pyautogui.locateCenterOnScreen("que.png", confidence=0.7))
    q = list(pyautogui.locateCenterOnScreen("pick.png", confidence=0.7))
    p[0] = 0.5*(p[0]+q[0])
    p[1] = 0.5*(p[1]+q[1])
    pyautogui.moveTo(p)
    pyautogui.click(clicks=3, interval=0.2)
    pyautogui.hotkey('ctrl', 'c')
    data = pyperclip.paste()
    data = clean_str(data)
    return data
```

```python
def click_the_buttom(filename):
    p = pyautogui.locateCenterOnScreen(filename, confidence=0.7)
    pyautogui.moveTo(p[0],p[1],1.2,pyautogui.easeOutQuad)
    pyautogui.click()
    time.sleep(2)
    

def save_page(filename):
    click_the_buttom('question.png')
    time.sleep(1)
    pyautogui.hotkey('ctrl', 's')
    time.sleep(0.5)
    pyautogui.write(filename, interval=0.2)
    pyautogui.press('enter')
    time.sleep(1.5)
    #in case of overwriting
    pyautogui.press('left')
    pyautogui.press('enter')
    time.sleep(5)
```
